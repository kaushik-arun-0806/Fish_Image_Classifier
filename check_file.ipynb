```python
import h5py

input_file = "best_fish_classifier.h5"
output_file = "best_fish_classifier_fixed.h5"

with h5py.File(input_file, "r") as src:
    with h5py.File(output_file, "w") as dst:

        # Copy everything
        src.copy("/", dst)

        # Read model config
        config = src.attrs["model_config"]

        if isinstance(config, bytes):
            config = config.decode("utf-8")

        # Remove the problematic field
        config = config.replace(
            ', "quantization_config": null',
            ''
        )

        dst.attrs["model_config"] = config

print("Fixed model saved as:", output_file)
```
